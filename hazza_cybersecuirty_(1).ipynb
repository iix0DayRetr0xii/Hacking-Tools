 **Homework : Mini Authentication Project**

Objective: Simulate a real login system.

*  **Scenario:**
One fixed username

One fixed password

The user:

Enters username and password

Has 3 attempts

If login succeeds:

Print Welcome to Dashboard

If login fails:

Print Access Denied

*   **Bonus (Advanced)**:

Display remaining attempts.

Hide password input (research task).

```python
username = "admin"
password = "1234"

attempts = 3

while attempts > 0:
    user_input = input("Enter username: ")
    pass_input = input("Enter password: ")

    if user_input == username and pass_input == password:
        print("Welcome to Dashboard")
        break
    else:
        attempts = attempts - 1
        print("Access Denied")
        print("Remaining attempts:", attempts)

if attempts == 0:
    print("Login failed. Try again later.")
```
